##### 2. X veri kümesi için aşağıdaki işlemleri gerçekleştiriniz. 
	a) Veri kümesinin her bir niteliği için ayrı ayrı [0-1] aralığına min-max normalizasyon ve z-score normalizasyon değerlerini hesaplayınız.
	b) Ayrık olmayan nitelik değerlerini (continuous attributes) n eşit genişlik (n equal-width) bölme yöntemleri ile ayrık hale dönüştürünüz (n değerini kendiniz belirleyebilirsiniz). Her bir yöntem sonucu elde edilen verilerin her bir niteliği için frekans tabanlı histogram grafiklerini çiziniz. Veri kümenizde belirtilen tüm nitelikler ayrık (discrete) ise bu seçeneği ihmal ediniz. Sadece sürekli değerli nitelikler için gerçekleştirilecektir.


##### Kütüphaneler eklendi.

```python
import pandas as pd
```

```python
from sklearn import preprocessing
```

```python
import numpy as np
```

```python
from scipy import stats
```

##### İlk nitelik için CSV okuma işlemi yapıldı.

```python
Attribute1=pd.read_csv("SmoothnessReal.csv")
Attribute1
```

##### İlk nitelik için Min-Max Normalizasyon işlemi yapıldı.

```python
x = Attribute1.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
Attribute1 = pd.DataFrame(x_scaled)
Attribute1
```

#####  İlk nitelik için z-score normalizasyon değerleri

```python
ZScore1=stats.zscore(Attribute1)
ZScore1
```

##### İlk nitelik için ayrık olmayan nitelik değerlerini (continuous attributes) n eşit genişlik (n equal-width) bölme yöntemleri ile ayrık hale dönüştürüldü. n 4 alındı

```python
DivisionAttribute1 = pd.cut(Attribute1.Smoothness1_real, bins=4, labels=np.arange(4), right=False)
DivisionAttribute1
```

##### İlk nitelikte elde edilen verilerin her bir niteliği için frekans tabanlı histogram grafiklerini çizildi.

```python
DivisionAttribute1.hist()
```
